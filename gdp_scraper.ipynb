```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36"
}
url_usa = "https://www.macrotrends.net/global-metrics/countries/USA/united-states/gdp-per-capita"
usa_requests = requests.get(url_usa, headers=headers)
pd.set_option('display.max_rows', None)
```

```python
usa_requests
```

```python
usa_soup = BeautifulSoup(usa_requests.content, "html.parser")
```

```python
usa_soup
```

```python
usa_table = usa_soup.find_all("table", class_="historical_data_table table table-striped table-bordered")[-1]
```

```python
usa_data = []
for row in usa_table.find_all("tr")[1:]:
    cols = row.find_all("td")
    if len(cols) >= 3:
        year = cols[0].text.strip()
        gdp_per_capita = cols[1].text.strip().replace("$", "").replace(",", "")
        annual_growth_rate = cols[2].text.strip().replace("%", "")
        usa_data.append({"Year": year, "GDP Per Capita (US $)": gdp_per_capita, "Annual Growth Rate (%)": annual_growth_rate})
    else:
        print("Skipping row with insufficient columns:", row)

usa_df = pd.DataFrame(usa_data)
```

```python
url_italy = "https://www.macrotrends.net/global-metrics/countries/ITA/italy/gdp-per-capita"
italy_requests = requests.get(url_italy, headers=headers)
```

```python
italy_requests
```

```python
italy_soup = BeautifulSoup(italy_requests.content, "html.parser")
```

```python
italy_table = italy_soup.find_all("table", class_="historical_data_table table table-striped table-bordered")[-1]
```

```python
italy_data = []
for row in italy_table.find_all("tr")[1:]:
    cols = row.find_all("td")
    if len(cols) >= 3:
        year = cols[0].text.strip()
        gdp_per_capita = cols[1].text.strip().replace("$", "").replace(",", "")
        annual_growth_rate = cols[2].text.strip().replace("%", "")
        italy_data.append({"Year": year, "GDP Per Capita (US $)": gdp_per_capita, "Annual Growth Rate (%)": annual_growth_rate})
    else:
        print("Skipping row with insufficient columns:", row)

italy_df = pd.DataFrame(italy_data)
```

```python
italy_df.set_index("Year", drop=True, inplace=True)
usa_df.set_index("Year", drop=True, inplace=True)
```

```python
all_gdp = italy_df.join(usa_df, on="Year", lsuffix="_italy", rsuffix="_states")
```

```python
all_gdp.to_csv("all_gdp.csv")
```
